```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm

def plot_gaussian_mixture_pdf(means, covars, weights):
    max_mean = np.max(means)
    min_mean = np.min(means)
    max_covar = np.max(covars)
    x = np.linspace(min_mean - 3 * max_covar, max_mean + 3 * max_covar, 1000)
    y = np.zeros_like(x)
    for mean, covar, weight in zip(means, covars, weights):
        y += weight * norm.pdf(x, mean, covar)
    plt.plot(x, y)
```

```python
plt.figure(figsize=(3,3))
plot_gaussian_mixture_pdf(means=np.array([-1.5,0,1.5]), covars=np.array([0.1,0.4,0.1]), weights=np.array([0.05,0.95,0.05]))
plt.xlabel('Slipping velocity')
plt.ylabel('Probability density')
plt.grid()
plt.savefig('distrib_of_slipping_v.pdf', bbox_inches='tight')
```
